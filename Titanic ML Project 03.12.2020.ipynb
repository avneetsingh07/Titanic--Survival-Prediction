```python
# Logistic Regression model - Classification Algorithm

# Train Test Split.....................

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(data.drop("Survived", axis=1),data["Survived"], test_size=0.30, random_state=101)
```

```python
# Training and prediction 
#pip install -U scikit-learn
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
logmodel= LogisticRegression()
logmodel.fit(X_train , y_train)
```

```python
y_prediction = logmodel.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logmodel.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_prediction)
print(confusion_matrix)
```
